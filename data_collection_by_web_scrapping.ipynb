## Scrapping a table from a website:

### Website: cancer.org.au; Link: https://wiki.cancer.org.au/skincancerstats/UV_radiation#UV_Index

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url ="https://wiki.cancer.org.au/skincancerstats/UV_radiation#UV_Index"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
soup
```

```python
# Obtain every title of columns with tag <th>
header = []
for i in soup.find_all('th'):
 title = i.text
 header.append(title)
```

```python
 headers = [x.replace('\n', '') for x in header]
```

```python
# Create a dataframe
mydata = pd.DataFrame(columns = headers)
```

```python
# Create a for loop to fill mydata
for j in soup.find_all('tr')[1:]:
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```

```python
mydata= mydata.apply(lambda s: s.str.strip())
```

```python
mydata
```

```python
mydata_long_format = pd.melt(mydata, id_vars='Location', value_vars=['Jan', 'Feb', 'Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec'],
             var_name='Month', value_name='UV_Index')
```

```python
mydata_long_format
```

```python
mydata_long_format.to_csv("UVR")
```
